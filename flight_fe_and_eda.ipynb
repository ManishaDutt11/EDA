```python
# feature engineering

df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Date'].astype(int)
df['Year']=df['Date'].astype(int)
```

```python
df.info()
```

```python
# drop date of journey

df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time']=df['Arrival_Time'].str.split(" ").str[0]
```

```python
df.head()
```

```python
df["Arrival_hour"]=df['Arrival_Time'].str.split(":").str[0]
df["Arrival_minute"]=df['Arrival_Time'].str.split(":").str[1]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_minute']=df['Arrival_minute'].astype(int)
```

```python
df.info()
```

```python
df.drop(['Arrival_Time'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df["Departure_hour"]=df['Dep_Time'].str.split(":").str[0]
df["Departure_minute"]=df['Dep_Time'].str.split(":").str[1]
```

```python
df['Departure_hour']=df['Departure_hour'].astype(int)
df['Departure_minute']=df['Departure_minute'].astype(int)
```

```python
df.info()
```

```python
df.drop(['Dep_Time'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, np.nan:1, '4 stops':4})
```

```python
df[df["Total_Stops"].isnull()]
```

```python
df.drop(['Route'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df["Duration_hour"]=df['Duration'].str.split(" ").str[0].str.split("h").str[0]
```

```python
df["Duration_minute"]=df['Duration'].str.split(" ").str[1]
```

```python
df["Duration_minute"]=df["Duration_minute"].str.split("m").str[0]
```

```python
df.head()
```

```python
df["Duration_minute"]=df["Duration_minute"].fillna(0)
```
